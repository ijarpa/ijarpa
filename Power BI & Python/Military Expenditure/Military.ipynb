# Practica limpieza de datos y analisis en Python

```python
# importar principales librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga de archivo que vamos a utilizar
df = pd.read_excel("./Military+Expenditure.xlsx")
```

```python
# revisión primeras 5 filas y columnas
df.head()

# tenemos columnas que estan vacias y debemos cambiar el titulo de las columnas por los datos de la fila numero 2
```

```python
df.info()
```

# Limpieza de filas y columnas vacias

```python
# borramos las primeras 2 filas y las primeras 3 columnas del dataframe
df = df.iloc[2:,3:]
```

```python
df.head()
```

# reemplazar head de las columnas

```python
# reemplazamos el nombre de las columnas por los datos de la primera fila
df.columns = df.iloc[0]
```

```python
# eliminamos la primera fila (que es el titulo de las columnas)
df = df[1:]
```

```python
df.head()
```

# rellenar Columna continente con Forward fill

```python
df["Continent"] = df["Continent"].ffill()
```

```python
df.info()
```

# Limpiar Country Code- Country

```python
df.head()
```

```python
Country_name = df["Country Code - Country"].str.split("-", expand=True)
```

```python
Country_name.head()
```

```python
Country_name.columns = ['Code Country', 'Country', 'Empty']
```

```python
df = pd.concat([df, Country_name], axis=1)
```

```python
df = df.drop(['Country Code - Country', 'Empty'], axis=1)
```
