```python
pd.pivot_table(df[0:10], index='Nationality', values='Age', aggfunc='mean').plot(kind='bar')
```

```python
grafico = pd.pivot_table(df, index=('Nationality'), values='Age', aggfunc='mean')
grafico.sort_values('Age', ascending=False, inplace=True)
grafico[0:10].plot(kind='bar', color='red')
plt.title('Grafico 1')
```

```python
plt.hist(grafico, bins=10)
```
