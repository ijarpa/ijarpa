```python
# 11 Prove: Assignment Milestone

# Showing Creativity and Exceeding Requirements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("./life-expectancy.csv")
```

```python
df.describe()
```

```python
df.hist('Life expectancy (years)')
```

```python
top_10 = pd.pivot_table(df, index='Entity', values='Life expectancy (years)', aggfunc='mean')
top_10.sort_values('Life expectancy (years)', ascending=False, inplace=True)
```

```python

label = top_10.index[0:10]
top_10[0:10].plot(figsize=(21,9))
plt.xticks(np.arange(10), labels=label, rotation=90)
plt.show()
```
